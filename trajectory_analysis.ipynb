```python
import numpy as np

filename = 'sample_7l.txt'
jnt_data = np.loadtxt(filename,delimiter='\t',unpack=False)

jnt_data = np.array(jnt_data.tolist())
```

```python
%config InlineBackend.figure_formats = ['svg']
import matplotlib.pyplot as plt

plt.plot(jnt_data)
plt.legend(['Jnt1', 'Jnt2', 'Jnt3', 'Jnt4', 'Jnt5' ,'Jnt6', 'Ext1', 'Ext2', 'Ext3'])
plt.show()
```

```python
jnt_ang = jnt_data
jnt_vel = np.gradient(jnt_ang, axis=0)
jnt_acc = np.gradient(jnt_vel,axis=0)
jnt_jerk = np.gradient(jnt_acc, axis=0)

plt.plot(jnt_vel)
plt.show()
plt.plot(jnt_acc)
plt.show()
plt.plot(jnt_jerk)
plt.show()
```
